```python
import pandas as pd
df = pd.read_csv('17APR_Prog_Interm-1 поток  Python, контрольная работа-оценки.csv')
df.head()
```

```python
df.tail()
```

```python
df.drop(182, inplace=True)
```

```python
df.info()
```

```python
df = df.apply(lambda x: x.str.replace(',','.'))
```

```python
df = df.apply(lambda x: x.str.replace('-','0'))
```

```python
df = df[df.columns[7:]].apply(pd.to_numeric)
```

```python
df
```

```python
df['Оценка/11,00'].hist(bins=20);
```

```python
import numpy as np
a = []
b = []
c = []
d = 1
for i in df.columns[1:]:
    a.append(np.sum(df[i] > 0)) 
    b.append(round(np.sum(df[i] > 0)/len(df),2))
    c.append(str(d))
    d += 1
```

```python
data = pd.DataFrame({'Номер вопроса':c,'Кол-во верных решений':a,'Доля верных решений':b})
data
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(16,10))
plt.bar(data['Номер вопроса'], data['Кол-во верных решений'])
ax.set_xlabel('Номер вопроса', fontsize = 15)
ax.set_ylabel('Кол-во верных решений', fontsize = 15);
```

```python
fig, ax = plt.subplots(figsize=(16,10))
plt.bar(data['Номер вопроса'], data['Доля верных решений'])
ax.set_xlabel('Номер вопроса', fontsize = 15)
ax.set_ylabel('Доля верных решений', fontsize = 15);
```
